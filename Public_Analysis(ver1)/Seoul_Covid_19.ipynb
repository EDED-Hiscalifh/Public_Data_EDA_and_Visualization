```python
import pandas as pd
Dominant_PATH = '../Public_Analysis/Datasets/'
doc = pd.read_csv(Dominant_PATH + 'Seoul_Covid_19.csv', 
                  encoding = 'utf-8')
doc = doc.dropna(subset = ['지역'])
doc = doc[['확진일', '지역']]
doc = doc[::-1].reset_index(drop=True)
Region = doc['지역']
Region_list = list(Region.unique())
Region_list
Dates = doc['확진일']
Dates_list = list(Dates.unique())

Seoul_final = pd.DataFrame({'Region' : Region_list})
for j in range(len(Dates_list)) : 
    doc_copy = doc[doc['확진일']==Dates_list[j]]
    doc_copy = doc_copy.reset_index(drop= True)
    day = pd.DataFrame({'Region' : Region_list, Dates_list[j] : 0})
    for i in range(len(doc_copy)):
        for ind in range(len(day)) : 
            if day.loc[ind, 'Region'] == doc_copy.loc[i, '지역'] : 
                day.loc[ind, Dates_list[j]] += 1
    Seoul_final = pd.merge(Seoul_final, day)
    
Seoul_final = Seoul_final.set_index('Region')
Seoul_final = Seoul_final.cumsum(axis = 1)
Seoul_final = Seoul_final.drop(['타시도', '기타'])
```

# Making function 

```python
import pandas as pd

def change_column_name(DataFrame) :
    New_columns = list()
    for i in range(len(DataFrame.columns)) :
        String = DataFrame.columns[i].split('.')[0:3]
        if len(String[1]) == 1:
            String[1] = '0' + String[1]
        else : 
            String[1]
    
        if String[0] == '20' :
            String += ['2020']
        else :
            String += ['2021']
        String = String[1:]
        list(String)
        String = '/'.join(String)
        New_columns.append(String)
    DataFrame.columns = New_columns
    return DataFrame

def create_data_frame(filename) : 
    doc = pd.read_csv(Dominant_PATH + filename, encoding = 'utf-8')
    
    doc = doc.dropna(subset = ['지역'])
    doc = doc[['확진일', '지역']]
    doc = doc[::-1].reset_index(drop=True)
    
    Region = doc['지역']
    Region_list = list(Region.unique())
    Dates = doc['확진일']
    Dates_list = list(Dates.unique())
    
    Seoul_final = pd.DataFrame({'Region' : Region_list})
    for j in range(len(Dates_list)) : 
        doc_copy = doc[doc['확진일']==Dates_list[j]]
        doc_copy = doc_copy.reset_index(drop= True)
        day = pd.DataFrame({'Region' : Region_list, Dates_list[j] : 0})
        for i in range(len(doc_copy)):
            for ind in range(len(day)) : 
                if day.loc[ind, 'Region'] == doc_copy.loc[i, '지역'] : 
                    day.loc[ind, Dates_list[j]] += 1
        Seoul_final = pd.merge(Seoul_final, day)
    
    Seoul_final = Seoul_final.set_index('Region')
    Seoul_final = Seoul_final.cumsum(axis = 1)
    Seoul_final = Seoul_final.drop(['타시도', '기타'])
    return Seoul_final
```

```python
Dominant_PATH = '../Public_Analysis/Datasets/'
Seoul_confirmed = create_data_frame('Seoul_Covid_19.csv')
Seoul_confirmed
Seoul_confirmed_final = change_column_name(Seoul_confirmed)
Seoul_confirmed_final
Seoul_confirmed_final.to_csv("../Public_Analysis/Datasets/Seoul_final.csv")
```
